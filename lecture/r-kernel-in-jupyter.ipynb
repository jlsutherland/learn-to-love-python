# Question 1

```r
n <- 6
t <- 4
nt <- n * t
```

## Question 2

```r
l.t <- matrix(1, nrow = t)
l.n <- matrix(1, nrow = n)
l.nt <- matrix(1, nrow = nt)
```

## Question 3

```r
I.n <- diag(1, n)
I.t <- diag(1, t)
```

## Question 4

```r
idums <- I.n %x% l.t
idums
```

# Question 5

```r
m.n <- t(idums) %*% idums
solve(m.n)
```

We find a symmetric matrix of ratios.

## Question 6

```r
tdums <- l.t %x% I.n
tdums
```

## Question 7

```r
m.t <- t(tdums) %*% tdums
solve(m.t)
```

We find a symmetric matrix of ratios that is identical, in this case, to the matrix found from inverting the cross-sectional unit dummies.

## Question 8

```r
tidums <- cbind(idums, tdums)
m.ti <- t(tidums) %*% tidums
solve(m.ti)
```

We find that we cannot invert the matrix because it is singular (the determinant of the matrix is zero). A determinant of zero suggests that the solution set contains solutions that aren't linearly independent. _I.e._, one or more of the dummy rows or columns is/are exactly expressible as a linear combination of all or some other its rows or columns. We know that this is the case, because we haven't dropped any of the dummies.

## Question 9

```r
tidums <- tidums[,-12]
tidums
```

## Question 10

```r
m.ti2 <- t(tidums) %*% tidums
solve(m.ti2)
```

Because the matrix is now invertible, we yield the inverted matrix. We find that we are able to invert the matrix because none of its rows or columns are exactly expressible as a linear combination of all or some other its rows or columns (because we dropped one!).

## Question 11

```r
tidums <- cbind(rep(1, nrow(tidums)), tidums)
solve(t(tidums) %*% tidums)
```

We find that we cannot invert the matrix because it is singular. It is singular because we included the intercept term, causing one or more of its vectors to be exactly expressible as a linear combination of all or some of its other vectors.

In order to solve the matrix, we must also drop one of the cross-sectional dummies.

```r
tidums <- tidums[,-7]
solve(t(tidums) %*% tidums)
```
